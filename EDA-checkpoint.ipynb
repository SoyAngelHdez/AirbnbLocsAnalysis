```python
def validar_distribucion_normal(data, umbral = 0.05):
    test_stat_normality, p_value_normality=stats.shapiro(data)
    print("Valor p :%.4f" % p_value_normality)
    if p_value_normality < umbral:
        print("Rechazar hipótesis nula >> Los datos NO tienen una distribución normal")
    else:
        print("Fallo en rechazar la hipótesis nula >> Los datos tienen una distribucion normal")
```

```python
def validar_homogeneidad_varianza(conjunto1, conjunto2):
    test_stat_var, p_value_var= stats.levene(conjunto1,conjunto2)
    print("Valor de p:%.4f" % p_value_var)
    if p_value_var <0.05:
        print("Rechazamos la hipotesis nula >> La varianza de las muestra es distinta.")
    else:
        print("Fallamos en rechazar la hipotesis nula >> La varianza de la muestra es la misma.")
```

```python
variables = ['price', 'availability_365', 'minimum_nights']
    
for variable in variables:
    statistic, p_value = stats.shapiro(CDMXlist[variable].iloc[:5000])
    alpha = 0.05  # Nivel de significancia
    
    print('Prueba de normalidad -', variable)
    print('Estadística de prueba:', statistic)
    print('Valor p:', p_value)
    
    if p_value > alpha:
        print('La muestra sigue una distribución normal.')
        parametric = True
    else:
        print('La muestra no sigue una distribución normal.')
        parametric = False
    
    groups = []
    for i in alcaldias:
        groups.append(CDMXlist.loc[CDMXlist['neighbourhood'] == i, variable])
        
    statistic, p_value = stats.levene(*groups)
    
    print('Prueba de homogeneidad de varianza -', variable)
    print('Estadística de prueba:', statistic)
    print('Valor p:', p_value)
    
    if p_value > alpha:
        print('Las muestras tienen varianzas iguales.')
    else:
        print('Las muestras tienen varianzas diferentes.')
    
    # Determinar el tipo de prueba a utilizar
    if parametric:
        print('Se requiere utilizar pruebas paramétricas en el análisis.')
    else:
        print('Se requiere utilizar pruebas no paramétricas en el análisis.')
    
    print('-' * 40)
```

### Aquí comienzan nuestras predicciones

```python
no_lines= len(CDMXlist)
CDMXlist['ResidentsWorkFromMexico'] = 1
```

```python
print(CDMXlist)
for index in range(no_lines):
    posibility = random.randint(1,3)
    if posibility == 3:
        CDMXlist.loc[index,'ResidentsWorkFromMexico'] = 0
    else:
        CDMXlist.loc[index,'ResidentsWorkFromMexico'] = 1
```

```python
for variable in variables:
    Laboran = CDMXlist[variable][CDMXlist['ResidentsWorkFromMexico'] == 1]
    NoLaboran = CDMXlist[variable][CDMXlist['ResidentsWorkFromMexico'] == 0]
    
    statistic, p_value = stats.kruskal(Laboran, NoLaboran)
    alpha = 0.05 
    
    print('Variable:', variable)
    print('H0:', "No hay diferencia significativa en el promedio de", variable, "entre los grupos.")
    print('H1:', "Hay una diferencia significativa en el promedio de", variable, "entre los grupos.")
    
    if p_value < alpha:
        print('Resultado: Se rechaza la hipótesis nula (H0).')
        print("Significa que la", variable, "tiene una diferencia significativa entre los grupos.")
    else:
        print('Resultado: No se puede rechazar la hipótesis nula (H0).')
        print("Significa que la", variable, "no tiene una diferencia significativa entre los grupos.")
```
